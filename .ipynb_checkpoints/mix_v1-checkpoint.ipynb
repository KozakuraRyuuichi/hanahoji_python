```python
import numpy as np
import cv2
import time
from src.hand_tracker import HandTracker
```

```python
# 鼻検知
# カスケードファイルの読み込み
cascade_path =  "OpenCV-detection-models-master/haarcascades/haarcascade_mcs_nose.xml"
cascade = cv2.CascadeClassifier(cascade_path)
# 囲み線の色を緑に設定
rectangle_color = (0, 255, 0)  

# 指検知
# 宿題：何を読み込んでるのか理解必要
window = "Hand Tracking"
palm_model_path = "models/palm_detection_without_custom_op.tflite"
landmark_model_path = "models/hand_landmark.tflite"
anchors_path = "models/anchors.csv"

# 投影の色を設定
point_color = (0, 255, 0)
connection_color = (255, 0, 0)
thickness = 2

# カメラの読み込み
cap = cv2.VideoCapture(0)
```

```python
ret, frame = cap.read()
# 精度向上のためinput画像を白黒化
gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
noses = cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=10, minSize=(30, 30))
for nose in noses:
    print(nose)
    
ret
```

```python
while(cap.isOpened()):
    # フレームの読み取り
    ret, frame = cap.read()
    # フレーム読めなかったら終了    
    if not ret: break

    #        8   12  16  20
    #        |   |   |   |
    #        7   11  15  19
    #    4   |   |   |   |
    #    |   6   10  14  18
    #    3   |   |   |   |
    #    |   5---9---13--17
    #    2    \         /
    #     \    \       /
    #      1    \     /
    #       \    \   /
    #        ------0-

    # 宿題：指の投影ポイントに番号を割り振る
    # タプルにしてみた
    connections = (
        (0, 1), (1, 2), (2, 3), (3, 4),
        (5, 6), (6, 7), (7, 8),
        (9, 10), (10, 11), (11, 12),
        (13, 14), (14, 15), (15, 16),
        (17, 18), (18, 19), (19, 20),
        (0, 5), (5, 9), (9, 13), (13, 17), (0, 17)
    )

    # 宿題：検知器の設定？    
    detector = HandTracker(
        palm_model_path,
        landmark_model_path,
        anchors_path,
        box_shift=0.2,
        box_enlarge=1.3
    )
    
    # 宿題：おそらくdetectorメソッドがRGB読み込みと思われる
    image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    points, _ = detector(image)
    if points is not None:
        for point in points:
            x, y = point # それぞれのpointのx座標、y座標を取得
            # input画像、中心座標、半径、色、線幅
            cv2.circle(frame, (int(x), int(y)), thickness * 2, point_color, thickness)
        # おそらく線の始端と終端のx座標とy座標
#         for connection in connections:
#             x0, y0 = points[connection[0]] # connection[0] = (x0, y0)
#             x1, y1 = points[connection[1]] # connection[0] = (x1, y1)
#             #　input画像、始端座標、終端座標、色、線幅
#             cv2.line(frame, (int(x0), int(y0)), (int(x1), int(y1)), connection_color, thickness)　
    
    # 精度向上のためinput画像を白黒化
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    noses = cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=10, minSize=(30, 30))

    if len(noses) > 0:
        for nose in noses:
            print(nose)
            cv2.rectangle(frame, tuple(nose[0:2]),tuple(nose[0:2] + nose[2:4]), rectangle_color, thickness)

#     ここから鼻と指のい座標がうんたらかんたらを書く
#     鼻と指の検知描画はしない（処理重くなるから）

            
#     # 描画
#     cv2.startWindowThread()
    cv2.imshow('nose', frame)
    
    if cv2.waitKey(5) & 0xFF == ord('q'):
        break

# # 終了処理
cap.release()
cv2.destroyAllWindows()
```
